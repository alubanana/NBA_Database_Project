```python
import pandas as pd
```

```python
players = pd.read_csv("players_2015.csv")
```

```python
players = players.rename(columns = {'Unnamed: 0':'player_id', 'Player':'player_name', 
                                   'Tm':'team_name', 'Pos':'player_position',
                                   'BLK':'block','TRB':'rebounds','AST':'assists','STL':'steals',
                                    '2P%':'twopoint_shot_percentage','3P%':'threepoint_shot_percentage'
                                    ,'PER':'points_per_game'
                                
                                    })
```

```python
players.columns
```

```python
Players = players[['player_id','player_name', 'player_position','team_name','threepoint_shot_percentage',
                  'twopoint_shot_percentage','rebounds', 'assists', 'steals', 'block','points_per_game']]
```

```python
Players #差height, weight, salary, start_year
```

```python
player_all = pd.read_csv('Players_all.csv')
```

```python
Players = Players.merge(player_all, how = 'left', left_on='player_name', right_on='Player')
```

```python
Players = Players.drop(['Unnamed: 0','Player','collage','born','birth_city','birth_state',
              'Unnamed: 8','Unnamed: 9'], axis = 1)
```

```python
Players
```

```python
player_start =  pd.read_csv('player_data.csv')
```

```python
player_start[['name','year_start']]
```

```python
Players = Players.merge(player_start[['name','year_start']],how = 'left', left_on = 'player_name',
                       right_on = 'name')
```

```python
salary = pd.read_html('https://github.com/AddisonGauss/NbaData2015-2016/blob/master/nbasalaries.csv')
```

```python
salary = salary[0]
```

```python
salary['salary'] = salary['SALARY'].apply(lambda x: int(x.replace('$','').replace(',','')))
```

```python
salary
```

```python
Players = Players.merge(salary[['salary','Player']],how = 'left', left_on = 'player_name',
                       right_on = 'Player')
```

```python
Players
```

```python
Players['salary'].fillna((Players['salary'].mean()), inplace=True)
```

```python
import math
Players['year_start'].fillna((math.floor(Players['year_start'].mean())), inplace=True)
```

```python
Players = Players.drop(['name', 'Player'], axis = 1)
```

```python
Players.rename(columns = {'year_start':'start_year'})
```
